```python
import pandas as pd
import numpy as np
import csv
```

```python
def ove_price_generator (csv_file_input, csv_file_output):
    '''
    function for generating overnight stock price
    based on the subtraction of current day opening price 
    and closing price on the day before.
    
    Input:
        TAQ file with all selected stocks 
    
    Output:
        return updated file with overnight price gaps
        NaN represents the first trading day which doesn't
        have overnight gap.
    '''
    stock_data = pd.read_csv(csv_file_input)
    overnight_gap = - stock_data.groupby('symbol')['CPrc'].shift(1) \
                    + stock_data['OPrc']
    stock_data['Ovn_Gap'] = round(overnight_gap, 2)
    stock_data.to_csv(csv_file_output)
    return
```

```python
ove_price_generator('sushan_daily.csv','sushan_daily_updated.csv')
```
